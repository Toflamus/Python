Create `Cat`, and `Person`

```python


class Cat(object):
    def __init__(self, name, age = 1) -> None:
        self.name = name
        self.age = age
    
    def eat(self) -> None:
        print(f'{self}, is eating')

    def play(self) -> None:
        print(f'{self}, is play')

    def sleep(self) -> None:
        print(f'{self}, is sleeping')

    def catching_mouse(self) -> None:
        print(f'{self}, is catching a mouse')

    def __str__(self):
        return f'Cat: Name {self.name}, age {self.age}'

class Person(object):
    def __init__(self, name, pets, age = 1) -> None:
        self.name = name
        self.age = age
        self.pets = pets
    
    def eat(self) -> None:
        print(f'{self}, is eating')

    def play(self) -> None:
        print(f'{self}, is play')

    def sleep(self) -> None:
        print(f'{self}, is sleeping')

    def __str__(self):
        return f'Name {self.name}, age {self.age}'
    
    def raise_pets(self):
        for pet in self.pets: 
            pet.eat()
            pet.play()
            pet.sleep()

    # We need to specify the animals and they have different methods
    def training_pets(self):
        for pet in self.pets:
            if isinstance(pet, Dog):
                pet.watch()
            elif isinstance(pet, Cat):
                pet.catching_mouse()
        
    # If there are so many animals, what would happen?
    # We need to unify the name of methods
         
    
d = Dog('Wof', 1)
c = Cat('Neco', 2)
p = Person('Master',[d,c],18)
p.raise_pets()
p.training_pets()
```

OOP: Inheritance

```python
class Animal:
    def __init__(self, name, age = 1) -> None:
        self.name = name
        self.age = age

    def eat(self) -> None:
        print(f'Name {self.name}, age {self.age}, is eating')

    def play(self) -> None:
        print(f'Name {self.name}, age {self.age}, is play')

    def sleep(self) -> None:
        print(f'Name {self.name}, age {self.age}, is sleeping')

class Dog(Animal):
    def __init__(self, name, age = 1) -> None:
        super(Dog, self).__init__(name,age)

    def work(self) -> None:
        print(f'Name {self.name}, age {self.age}, is watching')

class Cat(Animal):
    def __init__(self, name, age = 1) -> None:
        super().__init__(name,age)

    def work(self) -> None:
        print(f'{self}, is catching a mouse')

    def __str__(self):
        return f'Cat: Name {self.name}, age {self.age}'

class Person(Animal):
    def __init__(self, name, pets, age = 1) -> None:
        super().__init__(name,age)
        self.pets = pets

    def __str__(self):
        return f'Name {self.name}, age {self.age}'
    
    def raise_pets(self):
        for pet in self.pets: 
            pet.eat()
            pet.play()
            pet.sleep()

    def training_pets(self):
        for pet in self.pets:
            pet.work()
   
        
d = Dog('Wof', 1)
c = Cat('Neco', 2)
p = Person('Master',[d,c],18)
p.raise_pets()
p.training_pets()
```
